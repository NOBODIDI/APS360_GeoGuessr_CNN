### Build brand new test set

Using dataset4 as a base, first update the image list to keep the ones that are still actually in the folder, and then remove all images in dataset4 that are also in the 32k dataset. Then randomly sample 60 images per geocell from the remaining set to form the new test dataset that is completely unseen before.

```python
import os
import pandas as pd

root_path_32k = '/content/drive/MyDrive/APS360 - Deep Learning Project /Colab Notebooks/32k_train_val_test'
root_path_d4 = '/content/drive/MyDrive/APS360 - Deep Learning Project /Colab Notebooks/dataset4_28000'
images_d4 = pd.read_csv('/content/drive/MyDrive/APS360 - Deep Learning Project /Colab Notebooks/imagery_locations_dataset4.csv')

image_ids_32k = []
for (dirpath, dirnames, filenames) in os.walk(root_path_32k):
    image_ids_32k.extend([f.replace('.png', '') for f in filenames])

image_ids_d4 = []
for (dirpath, dirnames, filenames) in os.walk(root_path_d4):
    image_ids_d4.extend([f.replace('.png', '') for f in filenames])

# retain the images in dataset4 that are actually still in the folder
images_new = images_d4[images_d4['pano_id'].isin(image_ids_d4)].reset_index(drop=True)

# get the images in dataset4 that are NOT in the 32k dataset
images_new = images_new[~images_new['pano_id'].isin(image_ids_32k)].reset_index(drop=True)
print(len(images_new))

# randomly sample 60 images per geocell
images_new = images_new.groupby('geocell_id').sample(60).reset_index(drop=True)
print(len(images_new))

images_new.to_csv('/content/drive/MyDrive/APS360 - Deep Learning Project /Colab Notebooks/final_representative_tests/imagery_locations_test_new.csv', index=False)
```
